```python
def checkROIIntensity(fname, elm, coordinates, roiRatioLim, plotFig = True):
    elmmap, x_pos, y_pos = getElmMap(fname, elm)
    xidx = np.where((x_pos - coordinates[0])**2 < 0.005)[0][1]
    yidx = np.where((y_pos - coordinates[1])**2 < 0.005)[0][1]

    ss = np.diff(x_pos)[0]
    xpts = round(coordinates[2]/ss)
    ypts = round(coordinates[3]/ss)
    
    ysize, xsize = elmmap.shape
    yst = yidx-ypts//2
    yed = yidx+ypts//2
    xst = xidx-xpts//2
    xed = xidx+xpts//2
    
    if yst < 0: yst = 0
    if xst < 0: xst = 0
    if yed > ysize: yed = ysize
    if xed > xsize: xed = xsize
        
    roi = elmmap[yst:yed, xst:xed]
    kmeanMap = kmean_analysis(2, elmmap, 42, plotoption = False)
    
    if plotFig:
        fig = plt.figure(figsize=(5,5))
        plt.pcolor(x_pos[xst:xed], y_pos[yst:yed], roi, cmap='gray', shading = 'auto')
        plt.title('Extracted ROI for Fine Scan')
        
    roi_mean = np.nanmean(roi)
    masked_data = elmmap * kmeanMap[1]
    masked_data[masked_data == 0] = np.nan
    map_mean = np.nanmean(masked_data)

    print('ROI_mean: %.2f;\t Feature_mean: %.2f\t'%(roi_mean, map_mean))
    if (roi_mean/map_mean < roiRatioLim):
        print('ROI_mean / Feature_mean = %.2f, which is less than %.2f'%(roi_mean/map_mean, roiRatioLim))
        return 0
    else:
        return 1
```
